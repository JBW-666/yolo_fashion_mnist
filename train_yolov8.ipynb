```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
USER_NAME = "jbw"
ROOT_DIR = f"/home/{USER_NAME}/yolo_fashion_mnist"
CONFIG_PATH = f"{ROOT_DIR}/fashion_mnist.yaml"  # é…ç½®æ–‡ä»¶ä½ç½®
IMG_SIZE = 128  # å’Œç”Ÿæˆæ•°æ®é›†æ—¶çš„å°ºå¯¸ä¸€è‡´
```

```python
model = YOLO("yolov8n.pt")
print("YOLOv8æ¨¡å‹åŠ è½½å®Œæˆ!")
```

```python
results = model.train(
    data=CONFIG_PATH,  # é…ç½®æ–‡ä»¶è·¯å¾„
    epochs=10,         # è®­ç»ƒè½®æ•°ï¼ˆå°ç™½å…ˆç»ƒ10è½®ï¼Œçº¦5åˆ†é’Ÿï¼‰
    imgsz=IMG_SIZE,    # è¾“å…¥å›¾ç‰‡å°ºå¯¸
    batch=16,          # æ‰¹æ¬¡å¤§å°ï¼ˆæ˜¾å­˜å°å°±æ”¹8ï¼‰
    save=True,         # ä¿å­˜æœ€ä½³æ¨¡å‹
    project=ROOT_DIR,  # è®­ç»ƒç»“æœä¿å­˜æ ¹è·¯å¾„
    name="runs",       # è®­ç»ƒç»“æœæ–‡ä»¶å¤¹åï¼ˆæœ€ç»ˆè·¯å¾„ï¼šruns/detect/train/ï¼‰
    save_period=5      # æ¯5è½®ä¿å­˜ä¸€æ¬¡æ¨¡å‹
)
```

```python
# åŠ è½½æœ€ä½³æ¨¡å‹ï¼ˆè®­ç»ƒåè‡ªåŠ¨ä¿å­˜ï¼‰
best_model = YOLO(f"{ROOT_DIR}/runs/weights/best.pt")
# æµ‹è¯•ä¸€å¼ åˆæˆå›¾ç‰‡ï¼ˆé€‰ç¬¬67å¼ ï¼‰
test_img_path = f"{ROOT_DIR}/yolo_dataset/images/train/67.jpg"
# æ¨¡å‹æ¨ç†
results = best_model.predict(test_img_path, imgsz=IMG_SIZE)

# å¯è§†åŒ–æ¨ç†ç»“æœ
res = results[0]
# å¼¹å‡ºçª—å£æ˜¾ç¤ºå¸¦æ ‡æ³¨æ¡†çš„å›¾ç‰‡
res.show()
# ç»ˆç«¯æ‰“å°è¯†åˆ«ç»“æœ
print("\nè¯†åˆ«ç»“æœ:")
for box in res.boxes:
    cls_idx = int(box.cls[0])  # ç±»åˆ«ç´¢å¼•
    confidence = box.conf[0].item()  # å‡†ç¡®åº¦ï¼ˆ0-1ï¼‰
    cls_name = [
        "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
        "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
    ][cls_idx]
    print(f"- ç±»åˆ«ï¼š{cls_name} | å‡†ç¡®åº¦ï¼š{confidence:.4f}")
```
